```python
sampledata = pd.read_csv('datapoint.csv',header=None)
lst = (sampledata.iloc[0,:]).tolist()

def list_Dataframe(list):
    df = pd.DataFrame(columns=[np.arange(0,(len(list)//4))],index=[0,1,2,3])
    
    for i in range(0,((len(list)//4)),2):
        df.iloc[0,i]  =list[(i*4)]
        df.iloc[0,i+1]=list[(i*4)+1]
        df.iloc[1,i]  =list[(i*4)+2]
        df.iloc[1,i+1]=list[(i*4)+3]
        df.iloc[2,i]=list[(i*4)+4]
        df.iloc[2,i+1]=list[(i*4)+5]
        df.iloc[3,i]=list[(i*4)+6]
        df.iloc[3,i+1]=list[(i*4)+7]
    return df

L_D = list_Dataframe(list=lst)
L_D.to_csv("video_pts.csv",index=None,header=None) # Convert Data frame to csv
```
